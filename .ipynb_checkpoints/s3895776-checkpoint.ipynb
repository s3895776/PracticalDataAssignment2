## Research Goal: 
Create a model that can classify music from 4 different classes of music using the attributes of the music. 

Info (from https://archive.ics.uci.edu/ml/datasets/Turkish+Music+Emotion+Dataset):
<br>
The dataset is designed as a discrete model. 
<br>
Four classes in the dataset: happy, sad, angry, relax. 
<br>
Verbal and non-verbal music.
<br>
Different genres of Turkish music.
<br>
A total of 100 music pieces are determined for each class. 
<br>
There are 400 observations. 
<br>
Each observation is 30 seconds of the original song. 

### Observations 
With a small sample size of 100 pieces per class, a higher accuracy is more likely to overfit, while a low accuracy suggests poor performance. Both should be considered when evaluating the model. 
<br>
The songs themselves are 30 second samples of the original song, which we don't have access to. This shouldn't matter, but is something to consider if using the same features for a longer sample song. 

## Data retrieval


```python
path = "Datasets/Turkish Music Emotion Dataset Data Set/Acoustic Features.csv"

turkish = pd.read_csv(path, header = 0)
```

Attribute information <br>
Mel Frequency Cepstral Coefficients (MFCCs), https://learn.flucoma.org/reference/mfcc/ 
<br>
Describes the contour of the spectrum, meaning the boundary of music frequencies.  
<br>

Tempo, https://en.wikipedia.org/wiki/Tempo 
<br>
Describes the speed the music plays.
<br>

Chromagram, https://en.wikipedia.org/wiki/Chroma_feature 
<br>
Describes pitch in discrete values for 12 classes. 
<br>

Spectral features, definition of Spectral from Oxford Dictionary:
<br>
concerning the spectrum of the music frequencies. 
<br>

Harmonic features, https://en.wikipedia.org/wiki/Harmonic
<br>
a positive integer multiplied by a fundamental frequency. 
<br>

These features analyze the emotional content in music signals.
<br>
MIR toolbox is used for feature extraction.

The following give descriptive information for discrete values:
<br>
Boxplot, histogram, scatter plot, scatter matrix

<br>
- Boxplot visualizes many descriptive features 
- Histogram + density plot visualizes distribution 
- Scatter plot to measure linearity between two variables 
- Scatter matrix to plot all variables. 

```python
from pandas.plotting import scatter_matrix
```

```python
harmonic.columns
```

```python
columns = {
    harmonic.columns[0]:"mean",
    harmonic.columns[1]:"std",
    harmonic.columns[2]:"slope",
    harmonic.columns[3]:"PFreq",
    harmonic.columns[4]:"PAmp",
    harmonic.columns[5]:"PEntropy"
}
```

```python
harmonic.describe()
```

```python
renamed_harmonic = harmonic.rename(columns = columns)

scatter_matrix(renamed_harmonic, figsize = (8, 8), diagonal = "hist")
plt.show()
```

```python
plt.hist(harmonic[harmonic.columns[3]], bins = 50)
plt.show()
```

Harmonic change P frequency has jumps in values. 

Mean and std have a linear relationship. If the mean and the std of the harmonic change increase together, the range of harmonic change also increases. Therefore for some songs, the range of the harmonic change will either be small or large. 
We can compare the small and large ranges and label them with the class labels to find correlation.
<br>
Mean and std distribution is also similar, but mean is more right skewed. 


```python
# scatter plot of mean and std labelled by target value

temp = target.unique()

target_1 = turkish["Class"] == temp[0]
target_2 = turkish["Class"] == temp[1]
target_3 = turkish["Class"] == temp[2]
target_4 = turkish["Class"] == temp[3]

turkish2 = turkish.copy(deep=True)
 
turkish2.loc[target_1, "Class"] = 0
turkish2.loc[target_2, "Class"] = 1
turkish2.loc[target_3, "Class"] = 2
turkish2.loc[target_4, "Class"] = 3

colors_p = {0: 'orange', 1: 'yellow', 2: 'blue', 3: 'red'}

colors = [colors_p[c] for c in turkish2["Class"]]
```

```python
feature1 = harmonic.columns[0]
feature2 = harmonic.columns[1]

color_map = ["orange = " + str(temp[0]), "yellow = " + str(temp[1]), "blue = " + str(temp[2]),
            "red = " + str(temp[3])]


plt.scatter(x= turkish2[feature1], y = turkish2[feature2], c = colors)
plt.xlabel(feature1)
plt.ylabel(feature2)
plt.show()
print(color_map)
```

```python
# Box plot
fig, axs = plt.subplots(1, 2, sharey = True)


sns.boxplot(ax = axs[0], x=target, y=feature1, data= turkish)

sns.boxplot(ax = axs[1], x=target, y=feature2, data= turkish)

plt.show()
```

```python
# Box plot
feature = "_Tempo_Mean"
sns.boxplot(x=target, y=feature, data= turkish)
```
