#### feature selection

With a small sample size and large feature pool, each feature may not be informative. With hill climbing, we will select the best features.

```python
from sklearn.utils import shuffle
y = target
x = features

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0, test_size=0.2, train_size=0.8)

# use this for training
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, random_state = 0, train_size = 0.75, test_size = 0.25)

y = y_train

features_np_array = features.loc[x_train.index]
features_np_array = features_np_array.to_numpy()

print(features_np_array.shape)

new_Ind = []
cur_MaxScore = 0.0
col_num = features_np_array.shape[1]
col_Ind_Random = shuffle(range(0,col_num), random_state=2)
for cur_f in range(0, col_num):
    
    new_Ind.append(col_Ind_Random[cur_f])
    newData = features_np_array[:, new_Ind]
    
    x_train, x_val, y_train, y_val = train_test_split(newData, y, test_size=0.25, random_state=0)
    clf = KNeighborsClassifier(n_neighbours, weights='distance')
    fit = clf.fit(x_train, y_train)
    cur_Score = clf.score(x_val, y_val)
    
#     above or equal to get only better features. 
    if cur_Score <= cur_MaxScore:
        new_Ind.remove(col_Ind_Random[cur_f])
        
    else:
        cur_MaxScore = cur_Score
        print("Score with " + str(len(new_Ind)) + " selected features: " + str(cur_Score))

        
print(new_Ind)
features_clipped = features.iloc[:, new_Ind]
print(features_clipped.columns)
```

Because of the risk of overfitting, we want to retain a good score while simplifying the model so it can generalize well. We can stop at a certain amount of selected features we are satisfied with. For this selection, 10 features is suitable for the model. 

```python
x = features_clipped
y = target

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0, test_size=0.2, train_size=0.8)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, random_state = 0, train_size = 0.75, test_size = 0.25)
print(x_val.shape)
print(x_test.shape)
```

```python
clf = KNeighborsClassifier(n_neighbors = n_neighbours, weights = "distance", p = 1)
fit = clf.fit(x_train, y_train)
predicted = fit.predict(x_val)
print(confusion_matrix(y_val, predicted))
print(classification_report(y_val, predicted))
```

```python
clf = KNeighborsClassifier(n_neighbors = n_neighbours, weights = "distance", p = 1)
fit = clf.fit(x_train, y_train)
predicted = fit.predict(x_test)
print(confusion_matrix(y_test, predicted))
print(classification_report(y_test, predicted))
```

### Decision tree

Will perform similar steps for train, val and test split.

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

x = features
y = target

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0, test_size=0.2, train_size=0.8)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, random_state = 0, train_size = 0.75, test_size = 0.25)
x_val.shape
```

```python
clf = DecisionTreeClassifier(random_state = 0)
fit = clf.fit(x_train, y_train)
predicted = fit.predict(x_val)
print( confusion_matrix(y_val, predicted) )
print( classification_report(y_val, predicted) )
```

Parameters to tune: 
- max_depth: Limit the complexity of the tree. Reduce overfitting.
- min_samples_split: reduce overfitting by stopping the split when samples are below minimum amount. 
- min_samples_leaf: cut leaves if number samples in leaves are too small. 
- max_features: reduce number of features to use. Introduce "best" feature s only. 

```python
clf = DecisionTreeClassifier(max_features = "sqrt", random_state = 0)
fit = clf.fit(x_train, y_train)
predicted = fit.predict(x_val)
print( confusion_matrix(y_val, predicted) )
print( classification_report(y_val, predicted) )
```

```python
# max features tuned
max_features = 25
max_depth = 5
min_samples_split = 5
min_samples_leaf = 3
clf = DecisionTreeClassifier(max_features = max_features, random_state = 0)
fit = clf.fit(x_train, y_train)
predicted = fit.predict(x_val)
print( confusion_matrix(y_val, predicted) )
print( classification_report(y_val, predicted) )
```

```python
# max depth to reduce overfititng one sample of splits.
clf = DecisionTreeClassifier(max_features = max_features, max_depth = max_depth, random_state = 0)
fit = clf.fit(x_train, y_train)
predicted = fit.predict(x_val)
print( confusion_matrix(y_val, predicted) )
print( classification_report(y_val, predicted) )
```

Decision tree classifier performs better with limited depth. 

```python
# min samples per split to reduce splits on small number of samples.
# given 10 samples, we can accept the result without splitting 
min_samples_split = 10
clf = DecisionTreeClassifier(max_features = max_features, max_depth = max_depth, min_samples_split = min_samples_split,
                             random_state = 0)

fit = clf.fit(x_train, y_train)
predicted = fit.predict(x_val)
print( confusion_matrix(y_val, predicted) )
print( classification_report(y_val, predicted) )
```

min samples split has little effect.
